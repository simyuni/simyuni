</div>

Colab을 통해 인기 있는 Python 라이브러리를 최대한 활용하여 데이터를 분석하고 시각화할 수 있습니다. 아래 코드 셀에서는 <strong>Numpy</strong>를 사용하여 임의의 데이터를 생성하고 <strong>매트플롯립</strong>으로 이를 시각화합니다. 셀을 클릭하면 코드 수정을 바로 시작할 수 있습니다.

```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
# 영어교육 수업안 생성

class LessonPlan:
    def __init__(self, lesson_title, grade_level, objectives, materials, activities, assessment):
        self.lesson_title = lesson_title
        self.grade_level = grade_level
        self.objectives = objectives
        self.materials = materials
        self.activities = activities
        self.assessment = assessment

    def display_plan(self):
        print(f"Lesson Title: {self.lesson_title}")
        print(f"Grade Level: {self.grade_level}")
        print("Objectives:")
        for obj in self.objectives:
            print(f"- {obj}")
        print("Materials Needed:")
        for mat in self.materials:
            print(f"- {mat}")
        print("Activities:")
        for act in self.activities:
            print(f"- {act}")
        print("Assessment:")
        print(self.assessment)

# 수업안 예시 데이터
lesson_title = "Exploring Peter Pan"
grade_level = "3rd Grade"
objectives = [
    "Introduce vocabulary from 'Peter Pan'",
    "Discuss themes of adventure and friendship",
    "Engage in role-play based on the story"
]
materials = [
    "Copy of 'Peter Pan'",
    "Whiteboard and markers",
    "Art supplies for drawing"
]
activities = [
    "Read excerpts from 'Peter Pan' together",
    "Group discussion on characters and themes",
    "Create a drawing of your favorite scene"
]
assessment = "Observe student participation during discussions and activities."

# LessonPlan 객체 생성 및 출력
lesson_plan = LessonPlan(lesson_title, grade_level, objectives, materials, activities, assessment)
lesson_plan.display_plan()
```

```
# 영어 교육 수업 활동 생성

class Activity:
    def __init__(self, activity_title, description):
        self.activity_title = activity_title
        self.description = description

    def display_activity(self):
        print(f"Activity: {self.activity_title}")
        print(f"Description: {self.description}\n")

# 활동 목록 생성
activities = [
    Activity("Role-Play", "Students act out a scene from 'Peter Pan', taking on different characters."),
    Activity("Vocabulary Game", "Create a matching game with vocabulary words from the story."),
    Activity("Group Discussion", "Discuss the main themes of friendship and adventure in small groups."),
    Activity("Art Project", "Draw or create a small diorama of Neverland or a favorite scene."),
    Activity("Story Mapping", "Create a visual map of the story’s events and characters."),
]

# 활동 출력
print("Class Activities for 'Exploring Peter Pan'\n")
for activity in activities:
    activity.display_activity()
```

```
# AI 활용 수업 활동 생성

class AIActivity:
    def __init__(self, activity_title, description, ai_tool):
        self.activity_title = activity_title
        self.description = description
        self.ai_tool = ai_tool

    def display_activity(self):
        print(f"Activity: {self.activity_title}")
        print(f"Description: {self.description}")
        print(f"AI Tool: {self.ai_tool}\n")

# AI 활동 목록 생성
ai_activities = [
    AIActivity(
        "Story Generation",
        "Use an AI text generator to create a new adventure story featuring Peter Pan.",
        "OpenAI's ChatGPT"
    ),
    AIActivity(
        "Interactive Vocabulary Quiz",
        "Leverage an AI-powered quiz tool to assess students' vocabulary knowledge from the story.",
        "Quizizz or Kahoot!"
    ),
    AIActivity(
        "Character Analysis",
        "Analyze characters' traits using an AI tool that generates insights based on text input.",
        "ChatGPT or similar AI text analysis tools"
    ),
    AIActivity(
        "Art Creation",
        "Students can use an AI art generator to create illustrations for their favorite scenes from the story.",
        "DALL-E or Midjourney"
    ),
    AIActivity(
        "Speech Practice",
        "Utilize an AI speech recognition tool to help students practice reading excerpts from 'Peter Pan'.",
        "Google Speech-to-Text"
    )
]

# AI 활동 출력
print("Class Activities Utilizing AI for 'Exploring Peter Pan'\n")
for activity in ai_activities:
    activity.display_activity()
```

```
# 필요한 라이브러리 설치
!pip install matplotlib wordcloud nltk gradio

import matplotlib.pyplot as plt
from wordcloud import WordCloud
import nltk
from collections import Counter
import gradio as gr

# NLTK 다운로드
nltk.download('punkt')
nltk.download('stopwords')

# 텍스트 처리
def generate_wordcloud(text):
    words = nltk.word_tokenize(text.lower())
    word_freq = Counter(words)
    wordcloud = WordCloud(width=800, height=400).generate_from_frequencies(word_freq)

    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.savefig('wordcloud.png')
    return 'wordcloud.png'

# Gradio 인터페이스
wordcloud_interface = gr.Interface(
    fn=generate_wordcloud,
    inputs="textarea",
    outputs="image",
    title="Wordcloud Vocabulary Learning App",
    description="Input text to generate a word cloud highlighting frequent words."
)

wordcloud_interface.launch()
```

```
!pip install gradio gtts

import gradio as gr
from gtts import gTTS

def text_to_speech(text):
    tts = gTTS(text)
    tts.save("output.mp3")
    return "output.mp3"

# Gradio 인터페이스
tts_interface = gr.Interface(
    fn=text_to_speech,
    inputs="text",
    outputs="audio",
    title="Text to Speech Converter",
    description="Enter text to convert it into speech."
)

tts_interface.launch()
```

```
!pip install gradio

import gradio as gr

# 클로즈 질문 정의
cloze_questions = [
    {"question": "Alex [BLANK] for his daring spirit.", "answer": "was known"},
    {"question": "He [BLANK] upon an ancient map.", "answer": "stumbled"},
]

def check_answer(user_answer, question_index):
    correct_answer = cloze_questions[question_index]["answer"]
    if user_answer.strip().lower() == correct_answer.lower():
        return "Correct!"
    else:
        return f"Incorrect! The correct answer is: {correct_answer}"

# Gradio 인터페이스
cloze_interface = gr.Interface(
    fn=check_answer,
    inputs=["text", "number"],
    outputs="text",
    title="Cloze Question Checker",
    description="Fill in the blank to check your answer."
)

cloze_interface.launch()
```

```
!pip install gradio

import gradio as gr

# 이미지 및 설명 정의
images = ["image1.png", "image2.png", "image3.png"]
descriptions = ["First image description.", "Second image description.", "Third image description."]

def check_sequence(a, b, c):
    correct_order = ['image1', 'image2', 'image3']
    user_order = [a, b, c]
    if user_order == correct_order:
        return "Correct order!"
    else:
        return "Try again!"

# Gradio 인터페이스
sequence_interface = gr.Interface(
    fn=check_sequence,
    inputs=["text", "text", "text"],
    outputs="text",
    title="Image Sequencing Quiz",
    description="Enter the sequence of the images."
)

sequence_interface.launch()
```

```
!pip install transformers gradio

from transformers import pipeline
import gradio as gr

# 문법 검사기 로드
grammar_checker = pipeline("text2text-generation", model="prithivida/grammar_error_correcter_v1")

def grammar_check(text):
    corrected_text = grammar_checker(text)
    return corrected_text[0]['generated_text']

# Gradio 인터페이스
grammar_interface = gr.Interface(
    fn=grammar_check,
    inputs="text",
    outputs="text",
    title="Writing Activity: Grammar Checker",
    description="Provide text to check for grammar errors."
)

grammar_interface.launch()
```
